```python
import pandas as pd

data1=pd.read_csv("/content/Customers.csv")
display(data1)
data2=pd.read_csv("/content/Products.csv")
display(data2)
data3=pd.read_csv("/content/Transactions.csv")
display(data3)
```

Merge these datasets to create a unified dataset

```python
merge_data=data3.merge(data1,on="CustomerID",how="left") \
                .merge(data2,on="ProductID",how="left")
print(merge_data)
```

Key feature to know about the customers

```python
customer_summary = merge_data.groupby('CustomerID').agg(
    total_spent=('TotalValue', 'sum'),
    num_transactions=('TransactionID', 'count'),
    most_common_product=('ProductID', lambda x: x.mode()[0]),
    most_common_category=('Category', lambda x: x.mode()[0])
).reset_index()
print(customer_summary)
```

Using euclidean distance to calculate the similarity between customers. For this encode the categorical data value into integer incoding using sklearn library

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
customer_summary=merge_data.groupby('CustomerID').agg(
    total_spent=('TotalValue', 'sum'),
    num_transactions=('TransactionID', 'count'),
    most_common_product=('ProductID', lambda x: x.mode()[0]),
    most_common_category=('Category', lambda x: x.mode()[0])
).reset_index()
print(customer_summary)
```

```python
label_encoder = LabelEncoder()
customer_summary['most_common_product'] = label_encoder.fit_transform(customer_summary['most_common_product'])
customer_summary['most_common_category'] = label_encoder.fit_transform(customer_summary['most_common_category'])
print(customer_summary)
```

Ignoring irrelevant data values for calculating similarity score

```python
features = customer_summary.drop(columns=['CustomerID', 'most_common_product', 'most_common_category'])
```

```python
distance_matrix=euclidean_distances(features)
print(distance_matrix)
```

```python
similarity_matrix=1/(1+distance_matrix)
print(similarity_matrix)
```

```python
lookalike_map={}
for i in range(20):
  similarity_scores=similarity_matrix[i]
  similarity_customers=np.argsort(similarity_scores)[::-1][1:4]
  similar_customers_scores=[(customer_summary.iloc[j]["CustomerID"],similarity_scores[j]) for j in similarity_customers]
  lookalike_map[customer_summary.iloc[i]["CustomerID"]]=similar_customers_scores
  print(f"Customer {customer_summary.iloc[i]['CustomerID']}: {similar_customers_scores}")
```

Convert the data values in dataframe format

```python
lookalike_df=pd.DataFrame(
    [(key,val[0][0],val[0][1],val[1][0],val[1][1],val[2][0],val[2][1])
     for key,val in lookalike_map.items()],
     columns=['CustomerID','Lookalike1_ID','Similarity1','Lookalike2_ID','Similarity2','Lookalike3_ID','Similarity3']
)
display(lookalike_df)
```
