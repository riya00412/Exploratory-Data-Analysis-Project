Task 1: Exploratory Data Analysis (EDA) and Business Insights

```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
data1=pd.read_csv("/content/Customers.csv")
display(data1)
data2=pd.read_csv("/content/Products.csv")
display(data2)
data3=pd.read_csv("/content/Transactions.csv")
display(data3)
```

```python
data1.isnull().sum()
data1.dropna()
```

```python
data2.isnull().sum()
data2.dropna()
```

```python

data3.isnull().sum()
data3.dropna()
```

```python
duplicate_rows_df = data2[data2.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
duplicate_rows_df = data3[data3.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
data1.drop_duplicates(inplace=True)
print(data1.count)
```

```python
data2.drop_duplicates(inplace=True)
print(data2.count)
```

```python
data3.drop_duplicates(inplace=True)
print(data3.count)
```

```python
data1["SignupDate"]=pd.to_datetime(data1["SignupDate"])
print(data1.dtypes)
data3["TransactionDate"]=pd.to_datetime(data3["TransactionDate"])
print(data3.dtypes)
```

```python
# Total Customer
Total_customer=data1["CustomerID"].nunique()
print("Total Customer:",Total_customer)
```

```python
# Total Products
Total_products=data2["ProductID"].nunique()
print("Total Products:",Total_products)
```

```python
# Total Transection
Total_transactions=data3["TransactionID"].nunique()
print("Total Transection:",Total_transactions)
```

```python
merge_data=data3.merge(data1,on="CustomerID",how="left") \
                .merge(data2,on="ProductID",how="left")
print(merge_data)
```

```python
# Total Revenue
Total_revenue=merge_data["TotalValue"].sum()
print("Total Revenue:",Total_revenue)
```

```python
# Avg Transaction Value
Avg_transaction_value=merge_data["TotalValue"].mean()
print("Avg Transaction Value:",Avg_transaction_value)
```

```python
# Top Regions by Transactions
Top_regions=merge_data["Region"].value_counts().head(5)
print("Top 5 Regions:",Top_regions)
```

```python
# Top Categories by Sales
top_categories=merge_data.groupby("Category")["TotalValue"].sum().sort_values(ascending=False).head(3)
print("Top Categories by Sales:",top_categories)
```

```python
Top_regions.plot(kind="bar",color="red",title="Top Regions by Transactions")
plt.xlabel("Region")
plt.ylabel("Total Transactions")
plt.show()
```

```python
top_categories.plot(kind="pie",autopct='%1.1f%%',title="Top Categories by Sales")
plt.ylabel("")
plt.show()
```

```python
merge_data["TransactionMonth"]=merge_data["TransactionDate"].dt.to_period("M")
revenue_trend=merge_data.groupby("TransactionMonth")["TotalValue"].sum()
print(revenue_trend)
```

```python
revenue_trend.plot(kind="line",marker="o",title="Revenue Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```
